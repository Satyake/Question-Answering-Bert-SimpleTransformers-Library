```python
import json 
with open (r"train.json","r") as read_file:
  train=json.load(read_file)
```

```python
import json 
with open (r"test.json","r") as read_file:
  test=json.load(read_file)
```

```python
test
```

```python
from simpletransformers.question_answering import QuestionAnsweringModel,QuestionAnsweringArgs
```

```python
model_type="bert"
model_name="bert-base-cased"
if model_type=='bert':
  model_name='bert-base-cased'
elif model_type=="roberta":
  model_name="roberta-base"
elif model_type=="distilbert":
  model_name="distilbert-base-cased"

```

```python
!rm -rf /content/outputs
```

```python
#configure the model
model_args=QuestionAnsweringArgs()
model_args.train_batch_size=16
model_args.evaluate_during_training=True
model_args.n_best_size=3
model_args.num_train_epochs=30
```

```python
model=QuestionAnsweringModel(model_type,model_name,args=model_args)
```

```python
history=model.train_model(train,eval_data=test)
```

```python
result,text=model.eval_model(test)
```

```python
result
```

```python
text
```

```python
to_predict=[
    { 
    "context": "Vin is a Mistborn of great power and skill",
    "qas": [
        {
            "question": "What is Vins Speciality?",
            "id": "0"
        }
    ]
}]
```

```python
answers,probabilities=model.predict(to_predict)
```
